## Taylor Serier using Cosine

````python
def cos_taylor_series(x):
    tn = 0    # T(x)
    ft = 1    # factorial
    sign = 1  # sign = -1 or +1
    error = 0 # value for the end of the true loop
    n = 0     # n'th value

    while True:  #Python's do while loop syntax
        tn = tn + sign*(x**n)/ft  # T(x) value for nth term
        if abs(tn-np.cos(x)) < 10**(-12):  #Stop when n is found
            break
        error = np.cos(x)-tn # error calculatation    
        sign = sign*-1  # change sign for next term
        ft = ft*(n+1)*(n+2)  # factorial for next term
        n = n+2  # increase k by 2 
    print("x: {0}\nT(x): {1} \nExact: {2} \nSmallest n: {3} \nRemainder Term {4}\n".format(x, tn, np.cos(x), n, error))

for i in range(1, 5):  # for j=1,2,3,4
    cos_taylor_series(i/10)
    
````
